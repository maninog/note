```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.metrics import roc_auc_score
import xgboost as xgb
```

```python
df = pd.read_pickle("data_v0.pkl")
print(df.shape)
df.head(3)
```

```python
ans = []
rep = 1000
for i in range(0, rep, 1):
    df_train, df_test = train_test_split(df, test_size=0.2, shuffle=True, random_state=i, stratify=df['obj'])
    df_train = df_train.copy()
    df_test = df_test.copy()
    df_train['train_test'] = 1
    df_test['train_test'] = 0
    df_mrg = pd.concat([df_train, df_test], axis=0).reset_index(drop=True)
    #print(df.shape)

    x = df_mrg[df_mrg.columns[df_mrg.columns.str.startswith("F_")]]
    y = df_mrg['train_test']
    #print(x.shape, y.shape)
    #display(x, y)
    
    model = xgb.XGBClassifier(objective='binary:logistic', eval_metric='auc', max_depth=1, random_state=12345)
    model = model.fit(x, y)
    pred = model.predict_proba(x)[:, 1]
    ar = roc_auc_score(y, pred)*2-1
    #print(ar)
    ans.append([i, ar])

print(len(ans))
```

```python
arg_min = np.argmin(np.array(ans)[:,1])
print(np.argmin(np.array(ans)[:,1]))
print(np.min(np.array(ans)[:,1]))
print(np.array(ans)[arg_min])
```

```python
df_train, df_test = train_test_split(df, test_size=0.2, shuffle=True, random_state=arg_min, stratify=df['obj'])
df_train = df_train.copy()
df_test = df_test.copy()
df_train['train_test'] = 1
df_test['train_test'] = 0
df_mrg = pd.concat([df_train, df_test], axis=0).reset_index(drop=True)

x = df_mrg[df_mrg.columns[df_mrg.columns.str.startswith("F_")]]
y = df_mrg['train_test']

model = xgb.XGBClassifier(objective='binary:logistic', eval_metric='auc', max_depth=1, random_state=12345)
model = model.fit(x, y)
pred = model.predict_proba(x)[:, 1]
ar = roc_auc_score(y, pred)*2-1
print(ar)
```
